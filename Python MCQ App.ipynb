```python
def display_question(ques, ques_option_list):
    print(ques)
    for option in ques_option_list:
        print(option)
        
def check_answer(user_answer, correct_answer):
    return user_answer == correct_answer

def run_ques(Mcqs):
    score = 0
    
    for ques in Mcqs:
        display_question(ques["question"], ques["options"])
    
        user_answer = input("Chose Correct Answer: ")
    
        if check_answer(user_answer, ques["answer"]):
            print("Correct!")

            score += 1

        else:
            print(f"sorry your answer is wrong, correct answer is {ques['answer']}")
#             break
        
    print(f"your score is {score} out of {len(Mcqs)}")
        

Mcqs = [
    {
        "question": "What is the capital of Pakistan?",
        "options": ["Karchi", "Islamabad", "Peshawer"],
        "answer": "Islamabad"
    },

    {
        "question": "Whow is the Prime Minister of Pakistan?",
        "options": ["Shehbaz Sharif", "Imran Khan", "Asif Zardari"],
        "answer": "IMran Khan"
    },

    {
        "question": "Whow is the Founder of Shokat Khanam Hospital",
        "options": ["Imran Khan", "Musharraf", "Jweed Iqbal"],
        "answer": "Imran Khan"
    },

    {
        "question": "How much Population of Pakistan",
        "options": ["231.4 million", "201.4 million", "2301 million"],
        "answer": "231.4 million"
    },

    {
        "question": "Python is ______ languge",
        "options": ["Interpreted", "Compiler", "OOP", "both 1 & 3"],
        "answer": "both 1 & 3"
    }
]


run_ques(Mcqs)
```
